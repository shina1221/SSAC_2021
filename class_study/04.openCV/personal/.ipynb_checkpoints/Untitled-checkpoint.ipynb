```python
#그레이 스케일로 화면표시
img_file ='./img/girl.jpg'
img= cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)

if img is not None:
    cv2.imshow('IMG', img)
    cv2.waitKey()#키가 입력될때 까지 대기
    cv2.destroyAllWindows()# 키가 입력되면 제거 
else:
    print('No image file.')
```

```python
video_file = './img/big_buck.avi'

cap = cv2.VideoCapture(video_file)
if cap.isOpened():
    while True:
        ret, img =cap.read()
        print('ret',ret)
        if ret:
            #cv2.(윈도우 네임, mat)
            cv2.imshow(video_file, img)
            cv2.imshow?
            #waitKey() 는 키 입력을 기다리는 대기 함수입니다. 
            #매개변수로 넘긴 시간안에 키를 입력하면 소스의 다음줄로 이동
            cv2.waitKey(25)
        else:
            break
else:
    print("can't open video")
cap.release()
cv2.destroyAllWindows()
```
