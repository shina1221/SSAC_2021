```python
#FPS를 지정해서 동영상 재생
import cv2
video_file='C:/Users/admin/Downloads/img-20210913T062241Z-001/img/big_buck.avi'

#동영상 캠쳐 객제 생성
cap=cv2.VideoCapture(video_file)
#캠처 객체 초기화 확인
if cap.isOpened():
    #프레임 수 구하기 : CAP_PROP_FPS 함수로 FPS를 구할 수 있음 (초당 프레임 수)
    fps = cap.get(cv2.CAP_PROP_FPS)
    delay= int(1000/fps)
    print("FPS: %f, Delay: %dms"%(fps, delay))
    
    while True:
        ret, img = cap.read()
        if ret:
            cv2.imshow(video_file, img)
            cv2.waitKey(delay) #fps에 맞게 시간 지연
        else:
            break
            
else:
    print("can't open video")
cap.release()
cv2.destroyAllWindows()
```

```python
#창관리
import cv2

file_path = 'C:/Users/admin/Downloads/img-20210913T062241Z-001/img/girl.jpg'
img = cv2.imread(file_path)                            # 이미지를 기본 값으로 읽기
img_gray = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE) # 이미지를 그레이 스케일로 읽기

cv2.namedWindow('origin')                               # origin 이름으로 창 생성
cv2.namedWindow('gray', cv2.WINDOW_NORMAL)              # gray 이름으로 창 생성
cv2.imshow('origin', img)                               # origin 창에 이미지 표시
cv2.imshow('gray', img_gray)                            # gray 창에 이미지 표시

cv2.moveWindow('origin', 0, 0)                          # 창 위치 변경
cv2.moveWindow('gray', 100, 100)                        # 창 위치 변경

cv2.waitKey(0)                                          # 아무키나 누르면
cv2.resizeWindow('origin', 200, 200)                    # 창 크기 변경 (변경 안됨)
cv2.resizeWindow('gray', 100, 100)                      # 창 크기 변경 (변경 됨))

cv2.waitKey(0)                                          # 아무키나 누르면
cv2.destroyWindow("gray")                               # gray 창 닫기

cv2.waitKey(0)                                          # 아무키나 누르면
cv2.destroyAllWindows()                                 # 모든 창 닫기
```

```python
#플래그를 이용한 동그라미 그리기
import  cv2

title = 'mouse event'
img_file = 'C:/Users/admin/Downloads/img-20210913T062241Z-001/img/blank_500.jpg'
img = cv2.imread(img_file)
cv2.imshow(title, img)

colors = {'black':(0,0,0),
         'red':(0,0,255),
         'blue':(255,0,0),
         'green':(0,255,0)} #색상 미리정의


def onMouse(event, x, y ,flags, param):
    print(event, x, y, flags)
    color = colors['black']
    if event == cv2.EVENT_LBUTTONDOWN: #왼쪽 버튼을 누른 경우
        #컨트롤 키와 시프트 키를 모두 누른 경우
        if flags & cv2.EVENT_FLAG_CTRLKEY and flags & cv2.EVENT_FLAG_SHIFTKEY:
            color = colors['green']
        #시프트 키를 누른 경우
        elif flags & cv2.EVENT_FLAG_SHIFTKEY:
            color = colors['blue']
        #컨트롤 키를 누른 경우
        elif flags & cv2.EVENT_FLAG_CTRLKEY:
            color = colors['red']
        #지름 30크기의 검은원 좌표에 그림
        cv2.circle(img, (x,y), 30, color, -1)
        cv2.imshow(title, img)

#마우스 콜백 함수를 GUI 윈도에 등록
cv2.setMouseCallback(title, onMouse)
while True:
    #esc로 종료
    if cv2.waitKey(0) & 0xFF ==27:
        break
cv2.destroyAllWindows()
```

```python
#트랙바를 이용한 이미지 색 조정
import cv2
import numpy as np

#창이름
win_name ='Trackbar'

img_file = 'C:/Users/admin/Downloads/img-20210913T062241Z-001/img/blank_500.jpg'
img = cv2.imread(img_file)
#초기 이미지를 창에 표시
cv2.imshow(win_name, img)

#트랙바 이벤트 처리 함수 선언
def onChange(x):
    print(x) #트랙바의 새로운 위치값
    #'R', 'G', 'B' 각 트랙바 위치값
    r= cv2.getTrackbarPos('R', win_name)
    g= cv2.getTrackbarPos('G', win_name)
    b= cv2.getTrackbarPos('B', win_name)
    print(r,g,b)
    img[:] = [b,g,r] #기존 이미지에 win_name 픽셀 값 적용
    cv2.imshow(win_name, img) #새 이미지 창에 표시
    
#트랙바 생성
cv2.createTrackbar('R', win_name, 255,255, onChange)
cv2.createTrackbar('G', win_name, 255,255, onChange)
cv2.createTrackbar('B', win_name, 255,255, onChange)

while True:
    if cv2.waitKey(1) & 0xFF ==27:
        break
        
cv2.destroyAllWindows()
```
