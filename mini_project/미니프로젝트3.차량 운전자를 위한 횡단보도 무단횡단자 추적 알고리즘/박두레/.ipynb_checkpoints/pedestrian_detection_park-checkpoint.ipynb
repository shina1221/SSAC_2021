## 1. HOG-SVM 보행자 검출

### 1.1. HOG-SVM 보행자 검출 - default 디텍터

```python
# HOG-SVM 보행자 검출 - default 디텍터
import cv2
import imutils  # resize1

# default 디텍터를 위한 HOG 객체 생성 및 설정
# HOGDescriptor_getDefaultPeopleDetector() 함수에 의해서 제공되는 64 x 128 크기의 윈도로 훈련된 모델 이용
# HOGDescriptor() 생성자의 기본 값도 이와 같으므로 디폴트 생성자 사용
hogdef = cv2.HOGDescriptor()
hogdef.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

cap = cv2.VideoCapture('video1.mp4') 
# video1.mp4 / video2.gif / walking.avi

while cap.isOpened():
    ret, img = cap.read()
    if ret :
        # resize
        img = imutils.resize(img, width = 800)  # resize1
#         img = cv2.resize(img, dsize = None, fx = 0.5, fy = 0.5)  # resize2

        # default 디텍터로 보행자 검출 
        found, _ = hogdef.detectMultiScale(img)
        for (x, y, w, h) in found:
            cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 255), 1)
        cv2.putText(img, 'HOG-SVM Detector: Default', (10, 50), cv2.FONT_HERSHEY_DUPLEX,1, (0, 255, 0),1)
        cv2.imshow('img', img)
        
        key = cv2.waitKey(1)
        if key == 27:  # ESC
            break
        if key == 32:  # SPACE BAR
            cv2.waitKey(-1)  # wait until any key is pressed
    else:
        break
        
cap.release()
cv2.destroyAllWindows()
```

## 2. MOG2 배경 제거를 이용한 보행자 검출

### 2.1. MOG2-1

```python
# MOG2-1 지브코비치 알고리즘(BackgroundSubtractorMOG2)을 이용한 보행자 검출
import cv2
import imutils  # resize1

cap = cv2.VideoCapture('video1.mp4') 
# video1.mp4 / video2.gif / walking.avi

# 배경 제거 객체 생성
fgbg = cv2.createBackgroundSubtractorMOG2()

while cap.isOpened():
    ret, img = cap.read() 
    if ret:
        # resize
        img = imutils.resize(img, width = 800)  # resize1
#         img = cv2.cv2.resize(img, dsize = None, fx = 0.5, fy = 0.5)  # resize2

        # 배경 제거 마스크 계산
        fgmask = fgbg.apply(img)

        # 레이블링 함수를 이용하여 바운딩 박스 표시
        cnt, _, stats, _ = cv2.connectedComponentsWithStats(fgmask)  
        for i in range(1, cnt):
            x, y, w, h, area = stats[i]
            if area < 150:
                continue
            cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 1)
        cv2.putText(img, 'MOG2-1', (10, 50), cv2.FONT_HERSHEY_DUPLEX,1, (0, 255, 0), 1)
        cv2.imshow('img', img)

        key = cv2.waitKey(1)
        if key == 27:  # ESC
            break
        if key == 32:  # SPACE BAR
            cv2.waitKey(-1)  # wait until any key is pressed
    else:
        break
        
cap.release()
cv2.destroyAllWindows()
```

### 2.3. MOG2-1 & MOG2-2

```python
# MOG2-1 & MOG2-2 지브코비치 알고리즘(BackgroundSubtractorMOG2)을 이용한 보행자 검출
import cv2
import imutils  # resize1
import numpy as np

cap = cv2.VideoCapture('video1.mp4') 
# video1.mp4 / video2.gif / walking.avi

# 배경 제거 객체 생성
fgbg = cv2.createBackgroundSubtractorMOG2()
fgbg2 = cv2.createBackgroundSubtractorMOG2(varThreshold=350)

while cap.isOpened():
    ret, img = cap.read()
    if img is None:
        break
    img2 = img.copy()
    if ret:
        # resize
        img = imutils.resize(img, width = 800)  # resize1
        img2 = imutils.resize(img2, width = 800)  # resize1
#         img = cv2.cv2.resize(img, dsize = None, fx = 0.5, fy = 0.5)  # resize2
#         img2 = cv2.cv2.resize(img, dsize = None, fx = 0.5, fy = 0.5)  # resize2

        # 배경 제거 마스크 계산
        fgmask = fgbg.apply(img)
        fgmask2 = fgbg2.apply(img2)

        # 1
        cnt, _, stats, _ = cv2.connectedComponentsWithStats(fgmask)  # cnt: 객체 수 + 1 (배경 포함)
        for i in range(1, cnt):  # 각각의 객체 정보에 들어가기 위한 반복문, 1부터 시작 - 배경 제외
            x, y, w, h, area = stats[i]  # 좌측 상단 좌표 x, y, width, height, area(면적, 픽셀 수)
            # 노이즈 제거
            if area < 150:
                continue
            cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255))
        cv2.putText(img, 'MOG2-1', (10, 50), cv2.FONT_HERSHEY_DUPLEX,1, (0,255,0),1)
        cv2.imshow('img', img)

        # 2
        cnt, _, stats, _ = cv2.connectedComponentsWithStats(fgmask2)  
        for i in range(1, cnt):
            x, y, w, h, area = stats[i]
            if area < 100:
                continue
            cv2.rectangle(img2, (x, y), (x + w, y + h), (255, 0, 255), 1)
        cv2.putText(img2, 'MOG2-2', (10, 50), cv2.FONT_HERSHEY_DUPLEX,1, (0,255,0),1)
        cv2.imshow('img2', img2)
        
        # 3
        cv2.imshow('bgsub', fgmask)
        
        key = cv2.waitKey(1)
        if key == 27:  # ESC
            break
        if key == 32:  # SPACE BAR
            cv2.waitKey(-1)  # wait until any key is pressed
    else:
        break
        
cap.release()
cv2.destroyAllWindows()
```

### 3. 보행자 검출 전체

```python
import cv2
import imutils  # resize1
import numpy as np

# default 디텍터를 위한 HOG 객체 생성 및 설정
# HOGDescriptor_getDefaultPeopleDetector() 함수에 의해서 제공되는 64 x 128 크기의 윈도로 훈련된 모델 이용
# HOGDescriptor() 생성자의 기본 값도 이와 같으므로 디폴트 생성자 사용
hogdef = cv2.HOGDescriptor()
hogdef.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

# daimler 디텍터를 위한 HOG 객체 생성 및 설정
# HOGDescriptor_getDaimlerPeopleDetector() 함수가 제공하는 48 x 96 크기의 윈도로 훈련된 모델을 이용하므로 생성자도 이에 맞게 설정
hogdaim = cv2.HOGDescriptor((48,96), (16,16), (8,8), (8,8), 9)
hogdaim.setSVMDetector(cv2.HOGDescriptor_getDaimlerPeopleDetector())

cap = cv2.VideoCapture('video1.mp4') 
# video1.mp4 / video2.gif / walking.avi

# 배경 제거 객체 생성
fgbg = cv2.createBackgroundSubtractorMOG2()
fgbg2 = cv2.createBackgroundSubtractorMOG2(varThreshold=350)

while cap.isOpened():
    ret, img = cap.read()
    if img is None:
        break
    img2 = img.copy()
    img3 = img.copy()
    img4 = img.copy()
    if ret :
        # resize
        img = imutils.resize(img, width = 800)  # resize1
        img2 = imutils.resize(img, width = 800)  # resize1
        img3 = imutils.resize(img, width = 800)  # resize1
        img4 = imutils.resize(img, width = 800)  # resize1
#         img = cv2.cv2.resize(img, dsize = None, fx = 0.5, fy = 0.5)  # resize2
#         img2 = cv2.cv2.resize(img, dsize = None, fx = 0.5, fy = 0.5)  # resize2
#         img3 = cv2.cv2.resize(img, dsize = None, fx = 0.5, fy = 0.5)  # resize2
#         img4 = cv2.cv2.resize(img, dsize = None, fx = 0.5, fy = 0.5)  # resize2

        # default 디텍터로 보행자 검출 
        found, _ = hogdef.detectMultiScale(img)
        for (x, y, w, h) in found:
            cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 255), 1)
        cv2.putText(img, 'HOG-SVM Detector: Default', (10, 50), cv2.FONT_HERSHEY_DUPLEX,1, (0, 255, 0),1)
        cv2.imshow('img', img)
        
        # daimler 디텍터로 보행자 검출
        found2, _ = hogdaim.detectMultiScale(img2)
        for (x, y, w, h) in found2:
            cv2.rectangle(img2, (x, y), (x + w, y + h), (0, 255, 0), 1)
        cv2.putText(img2, 'HOG-SVM Detector: Daimler', (10, 50), cv2.FONT_HERSHEY_DUPLEX,1, (0, 255, 0),1)
        cv2.imshow('img2', img2)
        
        # 지브코비치 알고리즘(BackgroundSubtractorMOG2)을 이용한 보행자 검출
        # 배경 제거 마스크 계산
        fgmask = fgbg.apply(img3)
        fgmask2 = fgbg2.apply(img4)
        
        # 1
        cnt, _, stats, _ = cv2.connectedComponentsWithStats(fgmask)  
        for i in range(1, cnt):  
            x, y, w, h, area = stats[i]  
            if area < 150:
                continue
            cv2.rectangle(img3, (x, y), (x + w, y + h), (0, 0, 255))
        cv2.putText(img3, 'MOG2-1', (10, 50), cv2.FONT_HERSHEY_DUPLEX,1, (0,255,0),1)
        cv2.imshow('img3', img3)

        # 2
        cnt, _, stats, _ = cv2.connectedComponentsWithStats(fgmask2)  
        for i in range(1, cnt):
            x, y, w, h, area = stats[i]
            if area < 100:
                continue
            cv2.rectangle(img4, (x, y), (x + w, y + h), (255, 0, 255), 1)
        cv2.putText(img4, 'MOG2-2', (10, 50), cv2.FONT_HERSHEY_DUPLEX,1, (0,255,0),1)
        cv2.imshow('img4', img4)
        
        key = cv2.waitKey(1)
        if key == 27:  # ESC
            break
        if key == 32:  # SPACE BAR
            cv2.waitKey(-1)  # wait until any key is pressed
    else:
        break
        
cap.release()
cv2.destroyAllWindows()
```
