```python
#1개월 단위로 파생변수 뽑기
#3개월 단위로 파생변수 뽑기  

#현재 데이터로는 9개월 기준 3개월과 12개워 기준 3개월 값 밖에 알 수 없다.
#파생변수 산출 
#1월 값 -> 7월 기준 6개월 총액 - 6월 기준 6개월 총액- 7월 당월 값의 절댓값
#2월 값-> 8월 기준 6개월 총액 - 7월 기준 6개월 총액- 8월 당월 값 의 절댓값
#3월 기준 3개월(1~3월) -> 6월 기준 6개월 총액(1~6월) - 6월 기준 3개월 총액(4~6)
#                     = 6월 기준 6개월 총액 - ((12월 기준 12개월(1~12)총합 - 9월 기준 6개월(4~9) - 12월 기준 3개월(10~12)= 이하 4,5,6월 총액)  
#3월 값 -> 9월 기준 6개월 총액 - 8월 기준 6개월 총액- 9월 당월 값의 절댓값
#4월 값-> 10월 기준 6개월 총액 - 9월 기준 6개월 총액- 10월 당월 값 의 절댓값
#5월 값 -> 11월 기준 6개월 총액 - 10월 기준 6개월 총액 - 11월 당월 값 의 절댓값
#6월 기준 6개월(1~6월) -> 12개월 총액 - 12개월 기준 6개월 총액 = 6개월(1~6월) 총액 
#6월 값 -> 12월 기준 6개월 총액 - 11월 a기준 6개월 총액 - 12월 당워 값의 절댓값
#7월 기준 6개월(2~7월)
#8월 기준 6개월(3~8월)
#9월 기준 6개월(4~9월) -> 9월 기준 3개월 총액
#10월 기준 6개월(5~10월)
#11월 기준 6개월(6~11월)
#12월 기준 6개월(7~12월) -> 12월 기준 3개월 총액
```

```python
#6개월 총액 구하기(1월~6월)
#이용건수, 이용금액, 이용개월수 돌아가면서 만들어줄 것
pivot_df_12m = df[((df['기준년월'] == 201811) | (df['기준년월'] == 201812) &(df['이용건수_신판_R12M']>=0)&(df['이용금액_신판_R12M']>=0)) ].pivot(index='발급회원번호', columns = '기준년월', values = ['이용건수_신판_R12M','이용금액_신판_R12M','이용개월수_신판_R12M','이용건수_신판_R6M','이용금액_신판_R6M','이용개월수_신판_R6M'])
#6개월 총액 구하기(1월~6월)
##6월 기준 6개월(1~6월) -> 12개월 총액 - 12개월 기준 6개월 총액 = 6개월(1~6월) 총액 
#'이용건수_신판_R12M', '이용금액_신판_R12M', '이용개월수_신판_R12M'

#피벗테이블 되돌리기
chk_6m_1 = pivot_df_12m['이용건수_신판_R12M'][[201812]] - pivot_df_12m['이용건수_신판_R6M'][[201812]]
chk_6m_2 = pivot_df_12m['이용금액_신판_R12M'][[201812]] - pivot_df_12m['이용금액_신판_R6M'][[201812]]
chk_6m_3 = pivot_df_12m['이용개월수_신판_R12M'][[201812]] - pivot_df_12m['이용개월수_신판_R6M'][[201812]]

#세개의 피벗테이블 합치기 
chk_6m_1_melt = chk_6m_1.reset_index().melt(id_vars=['발급회원번호'])
chk_6m_2_melt = chk_6m_2.reset_index().melt(id_vars=['발급회원번호'])
chk_6m_3_melt = chk_6m_3.reset_index().melt(id_vars=['발급회원번호'])

#컬럼명 변경
chk_6m_1_melt.columns = ['발급회원번호','기준년월', '이용건수_신판_R6M'] #마지막 컬럼명 변경하면서 재활용
chk_6m_2_melt.columns = ['발급회원번호','기준년월', '이용금액_신판_R6M'] #마지막 컬럼명 변경하면서 재활용
chk_6m_3_melt.columns = ['발급회원번호','기준년월', '이용개월수_신판_R6M'] #마지막 컬럼명 변경하면서 재활용

#세 컬럼 합치기
plus_201806_6m = pd.merge(chk_6m_1_melt,chk_6m_2_melt[['발급회원번호','이용금액_신판_R6M']], on='발급회원번호', how='inner')
plus_201806_6m=pd.merge(plus_201806_6m,chk_6m_3_melt[['발급회원번호','이용개월수_신판_R6M']], on='발급회원번호', how='inner')

#기준년월 201806으로 변경
plus_201806_6m['기준년월'] = 201806
display(plus_201806_6m.head())

plus_201806_6m.describe()
#-값 존재 
#2018년 1월부터 6월까지의 총합이 -가 나오는 케이스도 존재하나 현재 단계에서는 지우지 않고 참고함. 
```

```python
#3개월 기준 피벗
#3개월 기준 총합(1~3월) = 6월 기준 6개월 총액 -  (4,5,6월 총액)
#                                          -  (12월 기준 12개월(1~12)총합 - 9월 기준 6개월(4~9) - 12월 기준 3개월(10~12)                                                   
#이용건수, 이용금액, 이용개월수 돌아가면서 만들어줄 것
#'이용건수_신판_R12M', '이용금액_신판_R12M', '이용개월수_신판_R12M'

#피벗테이블 만들기
#6월, 9월 추가
pivot_df_12m = df[((df['기준년월'] == 201809) | (df['기준년월'] == 201812) | (df['기준년월'] == 201806)) &(df['이용건수_신판_R12M']>=0)&(df['이용금액_신판_R12M']>=0) ].pivot(index='발급회원번호', columns = '기준년월', values = ['이용건수_신판_R12M','이용금액_신판_R12M','이용개월수_신판_R12M','이용건수_신판_R6M','이용금액_신판_R6M','이용개월수_신판_R6M','이용건수_신판_R3M','이용금액_신판_R3M','이용개월수_신판_R3M'])
#6월이 집계가 안되서 따로 진행
pivot_df_6m = df[(df['기준년월'] == 201806)].pivot(index='발급회원번호', columns = '기준년월', values = ['이용건수_신판_R6M','이용금액_신판_R6M','이용개월수_신판_R6M'])
#피벗테이블 되돌리기
#첫번째 컬럼 기준이 동일할 때 두번째 컬럼기준간 차이를 구하려면 [[]] 두개로 하고 
#첫번째 컬럼 기준이 다를 때 두번째 컬럼기준간 차이를 구하려면 [] 한개로 한다. 
#예시)pivot_df_12m['이용건수_신판_R6M'][[201812]] - pivot_df_12m['이용건수_신판_R6M'][[201809]]
#예시)pivot_df_12m['이용건수_신판_R6M'][201812] - pivot_df_12m['이용건수_신판_R6M'][201809]

#12월 기준 12개월 - 9월 기준 6개월
chk_456_1 = pivot_df_12m['이용건수_신판_R12M'][201812] - pivot_df_12m['이용건수_신판_R6M'][201809]
chk_456_2 = pivot_df_12m['이용금액_신판_R12M'][201812] - pivot_df_12m['이용금액_신판_R6M'][201809]
chk_456_3 = pivot_df_12m['이용개월수_신판_R12M'][201812] - pivot_df_12m['이용개월수_신판_R6M'][201809]

#12월 기준 12개월 - 9월 기준 6개월 - 12개월 기준 3개월
chk_456_1 = chk_456_1 - pivot_df_12m['이용건수_신판_R3M'][201812]
chk_456_2 = chk_456_2 - pivot_df_12m['이용금액_신판_R3M'][201812]
chk_456_3 = chk_456_3 - pivot_df_12m['이용개월수_신판_R3M'][201812]

#6개월 총합 - (4,5,6월) 총합
chk_456_1 = pivot_df_6m['이용건수_신판_R6M'][201806]-chk_456_1  
chk_456_2 = pivot_df_6m['이용금액_신판_R6M'][201806]-chk_456_2
chk_456_3 = pivot_df_6m['이용개월수_신판_R6M'][201806]-chk_456_3

#세개의 피벗테이블 melting
chk_456_1_melt = chk_456_1.reset_index().melt(id_vars=['발급회원번호'])
chk_456_2_melt = chk_456_2.reset_index().melt(id_vars=['발급회원번호'])
chk_456_3_melt = chk_456_3.reset_index().melt(id_vars=['발급회원번호'])
#일부 NaN값 존재. 이는 생성했던 2018년 6월 컬럼에 대한 6개월 총합데이터를 제외한 나머지 컬럼은 NaN값이기 때문에 
#원래 (df['이용건수_신판_R12M']>=0)&(df['이용금액_신판_R12M']>=0) 옵션을 사용할 수 없는 상태로 값을 산출했기 때문
#차후 이런 부분을 감안해서 0> 옵션을 걸어 집계를 진행필요

#컬럼명 변경
chk_456_1_melt.columns = ['발급회원번호','기준년월', '이용건수_신판_R3M'] #마지막 컬럼명 변경하면서 재활용
chk_456_2_melt.columns = ['발급회원번호','기준년월', '이용금액_신판_R3M'] #마지막 컬럼명 변경하면서 재활용
chk_456_3_melt.columns = ['발급회원번호','기준년월', '이용개월수_신판_R3M'] #마지막 컬럼명 변경하면서 재활용

#세 컬럼 합치기
plus_201803_3m = pd.merge(chk_456_1_melt,chk_456_2_melt[['발급회원번호','이용금액_신판_R3M']], on='발급회원번호', how='inner')
plus_201803_3m = pd.merge(plus_201803_3m,chk_456_3_melt[['발급회원번호','이용개월수_신판_R3M']], on='발급회원번호', how='inner')

#기준년월 201806으로 변경
plus_201803_3m['기준년월'] = 201803
#display(plus_201803_3m.head())

plus_201803_3m.describe()
#NaN, -값 존재하나 현재 단계에서는 지우지 않고 참고함. 
```
