# 2. 월별 사업장 수 데이터 분석 및 시각화

```python
data2018=[]
data2019=[]
data2020=[]
data2021=[]
```

### 2018년 월별 사업장 수 데이터 분석

```python
data2018=[]
for i in range(3,13):
    path=".\\data\\국민연금 가입 사업장 내역 2018년 %s월.csv"%str(i)
    data18=pd.read_csv(path,encoding="cp949",delimiter=',',low_memory=False)
    data18=data18.iloc[:,9:11]
    data2018.append(len(data18))
print(data2018)
```

```python
df2018={
    'business_number':[0,0]
}
df2018=pd.DataFrame(df2018)

for i in range(10):
    new_data={'business_number':data2018[i]}
    df2018=df2018.append(new_data,ignore_index=True)
df2018.index=["2018.1","2018.2","2018.3","2018.4","2018.5","2018.6","2018.7",\
              "2018.8","2018.9","2018.10","2018.11","2018.12"]
df2018
```

### 2019년 월별 사업장 수 데이터 분석

```python
data2019=[]
for i in range(1,13):
    path=".\\data\\국민연금 가입 사업장 내역 2019년 %s월.csv"%str(i)
    data19=pd.read_csv(path,encoding="cp949",delimiter=',',low_memory=False)
    data19=data19.iloc[:,9:11]
    data2019.append(len(data19))
print(data2019)
```

```python
df2019={
    'business_number':[data2019[0]]
}
df2019=pd.DataFrame(df2019)

for i in range(1,12):
    new_data={'business_number':data2019[i]}
    df2019=df2019.append(new_data,ignore_index=True)
df2019.index=["2019.1","2019.2","2019.3","2019.4","2019.5","2019.6","2019.7","2019.8","2019.9","2019.10","2019.11","2019.12"]
df2019
```

### 2020년 월별 사업장 수 데이터 분석

```python
data2020=[]
for i in range(1,13):
    path=".\\국민연금 가입 사업장 내역 2020년 %s월.csv"%str(i)
    data20=pd.read_csv(path,encoding="cp949",delimiter=',',low_memory=False)
    data20=data20.iloc[:,9:11]
    data2020.append(len(data20))
print(data2020)
```

```python
df2020={
    'business_number':[data2020[0]]
}
df2020=pd.DataFrame(df2020)

for i in range(1,12):
    new_data={'business_number':data2020[i]}
    df2020=df2020.append(new_data,ignore_index=True)
df2020.index=["2020.1","2020.2","2020.3","2020.4","2020.5","2020.6","2020.7","2020.8","2020.9","2020.10","2020.11","2020.12"]
df2020
```

### 2021년 월별 사업장 수 데이터 분석

```python
data2021=[]
for i in range(1,8):
    path=".\\국민연금 가입 사업장 내역 2021년 %s월.csv"%str(i)
    data21=pd.read_csv(path,encoding="cp949",delimiter=',',low_memory=False)
    data21=data21.iloc[:,9:11]
    data2021.append(len(data21))
print(data2021)
```

```python
df2021={
    'business_number':[data2021[0]]
}
df2021=pd.DataFrame(df2021)

for i in range(1,7):
    new_data={'business_number':data2021[i]}
    df2021=df2021.append(new_data,ignore_index=True)
for i in range(7,12):
    new_data={'business_number':0}
    df2021=df2021.append(new_data,ignore_index=True)
df2021.index=["2021.1","2021.2","2021.3","2021.4","2021.5","2021.6","2021.7","2021.8","2021.9","2021.10","2021.11","2021.12"]
df2021
```

### 월별 사업장 수 병합

```python
df_all=pd.concat([df2018,df2019,df2020,df2021])
df_all
```

###  전체 기간 사업장 수 변화 추이 시각화

```python
#df_all.iloc[2:-6,].plot()
plt.figure(figsize=(25,10))
plt.title('Business Number Transition by COVID19',fontsize=30)
plt.xlabel('2018 ~ 2021',fontsize=30)
plt.ylabel('number',fontsize=30)
plt.xticks(fontsize=10)
plt.yticks(fontsize=20)
plt.plot(df_all.iloc[2:-5,])
#plt.show()
plt.tight_layout()
plt.savefig('whole_period_business_number.png')
```
