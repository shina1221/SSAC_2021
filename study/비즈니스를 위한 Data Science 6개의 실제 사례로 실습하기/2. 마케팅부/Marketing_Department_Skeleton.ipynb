# TASK #5: FIND THE OPTIMAL NUMBER OF CLUSTERS USING ELBOW METHOD

```python
# Let's scale the data first
# 데이터 표준화 (표준화는 정규화와 다르게 반드시 0~1사이의 값을 보이지 않음)
scaler = StandardScaler() #= (x - mean())/std() //분산은1 평균은 0으로
credit_df_scaled = scaler.fit_transform(credit_df)
credit_df_scaled

"""
정규화된 데이터 스케일링 필요
BALANCE 열은 수천달러까지 값이 있지만
FREQUENCY 열은 0에서 1사이의 값만 존재
일시불 구매 또한 0에서 1사이의 값을 가짐

!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!1
만약 데이터를 스케일링 하지 않으면,
범위값이 큰 특성이 다른 특성들을 묻히게 만들것임

예를들면 FREQUENCY 같은 특성들
그렇기 때문에 항상 알고리즘을 적용하기 전 데이터를 스케일링 해야함.
"""
```

```python
credit_df_scaled.shape
```

```python
#elbow 메소드 
score_1 = []

range_vales= range(1,20)

for i in range_vales:
    kmeans = KMeans(n_clusters=i) #군집수 설정
    kmeans.fit(credit_df_scaled)
    score_1.append(kmeans.inertia_) #kmeans.inertia_
                                    #Inertia는 각 클러스터의 중심인 centroid와 각 데이터들 사이의 거리
                                    #Inertia가 낮은 그룹을 ‘좋은'그룹이라 할 수 있고, 이러한 그룹을 적게 만들수록 좋은 모델   
"""
시간이 비교적 좀 걸림

k값에 따른 모수 inertia가 바로 군집내 제곱합이 됨
현재 코드에서는 그 값과 군집의 수를 플롯 
"""
```

```python
plt.plot(score_1, 'bx-')
plt.title('finding the right number of clusters')
plt.xlabel('Clusters')
plt.ylabel('scores WCSS') #WCSS(Inside-Cluster-Sum-of-Squares) : 군집내 제곱합

plt.show()

# From this we can observe that, 4th cluster seems to be forming the elbow of the curve. 
# However, the values does not reduce linearly until 8th cluster. 
# Let's choose the number of clusters to be 7 or 8.
```

```python
"""
오토인코더는 일종의 인공신경망으로
데이터 인코딩 작업을 수행하기 위해 사용됨

오토인코더를 사용해 병목을 만듦.
파이프라인의 중간에 병목을 만들어 인풋데이터와 똑같은 아웃풋 데이터를 재구성함
오토인코더를 훈련시킬 때 실제로 동일한 인풋과 동일한 아웃풋 데이터를 사용
이는 비지도 학습방식에 속하며 주성분 분석과 비슷

인풋이 들어가 축소되는 것임
ex)고양이 이미지를 인풋과 아웃풋에 넣으면
   네트워크가 이 이미지를 인코딩해서 차원을 축소하고 압축하면서 정보가 소실되지 않도록 함.
   즉 용량을 줄임
   그리고 코드 층에 다다르면 이 데이터를 다시 업샘플링 하여 이미지를 동일하게 재구성
"""
```

```python
"""
모든 인코더는 네트워크에 병목을 추가해 작동
이 병목은 네트워크가 인풋을 압축시켜 인코딩되도록 강제함.

인코더는 인풋 데이터 간에 관계가 있을수록 잘 작동
만약 데이터에 상관관계들이 존재한다면 주성분 분석과 마찬가지로 오토인코더 또한 잘 작동함.

하지만 반대로 데이터가 독립적이어서 아무런 관계성이 없다면 오토인코더는 잘 작동하지 못하고, 수행결과도 안 좋음.
"""
```

```python
"""
인풋 X를 뉴런과 뉴런사이의 가중치로 곱 (W*X)
여기에 bias signal인 b를 더함. 이 값은 함수를 위 또는 아래로 이동시키기 위해 사용

이후 여기에 활성함수 적용
이 경우에는 시그모이드를 활성함수로 사용

h(x)값은 인풋을 인코딩한 값
즉, 인풋 x값이 들어오면 인코딩되어 h(x)가 된다는 것
이후 h(x)를 업샘플링 하는데 여기서도 가중치로 곱해줌
여기는 W*로 디코더의 가중치를 표현

bias signal또한 더해주는데 여기서는 C로 표현
그리고 시그모이드 함수를 적용한 최종 값은 X

만약 인코더가 잘 작동했다면 이 최종값은 처음 x값과 매우 비슷할 것임.

즉, 병목을 통과한 후 인풋값을 재구성.
인풋과 은닉층 사이의 가중치 
즉, 가중치 W와 가중치 W*는 동일한 값을 가짐
이것을 tied 가중치라고 함.(인코더 내의 가중치와 디코더 내의 가중치가 동일할 경우)
"""
```

# TASK #9: APPLY AUTOENCODERS (PERFORM DIMENSIONALITY REDUCTION USING AUTOENCODERS)

```python
from tensorflow.keras.layers import Input, Add, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, AveragePooling2D, MaxPooling2D, Dropout
from tensorflow.keras.models import Model, load_model

from tensorflow.keras.initializers import glorot_uniform 
#glorot_uniform 초기화 함수는 (하비에르)Xavier normal 초기화 함수라고도 함. 절단형 정규분포 데이터들에서 표본들을 뽑아냄

from tensorflow.keras.optimizers import SGD #확률적 경사 하강법
```

```python
input_df = Input(shape = (17,)) #17개의 특성 입력

x= Dense(7,activation='relu')(input_df) #Dense는 완전히 연결된 인공신경망을 의미
                                         #차원을 7이나 8로 지정
                                         #여기선 인코딩 차원은 7이 될 것임   
x = Dense(500, activation='relu', kernel_initializer='glorot_uniform')(x)
x = Dense(500, activation='relu', kernel_initializer='glorot_uniform')(x)
x = Dense(2000, activation='relu', kernel_initializer='glorot_uniform')(x)

encoded = Dense(10, activation='relu', kernel_initializer='glorot_uniform')(x) #뉴런 10개

x = Dense(2000, activation='relu', kernel_initializer='glorot_uniform')(encoded)
x = Dense(500, activation='relu', kernel_initializer='glorot_uniform')(x)

decoded = Dense(17, kernel_initializer='glorot_uniform')(x) 

#Autoencoder
autoencoder = Model(input_df, decoded)

#Encoder
encoder = Model(input_df, encoded)

autoencoder.compile(optimizer='adam', loss='mean_squared_error')
```
