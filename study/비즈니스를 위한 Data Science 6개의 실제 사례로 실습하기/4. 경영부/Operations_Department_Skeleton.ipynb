# TASK #6: IMPORT MODEL WITH PRETRAINED WEIGHTS

```python
#사전 학습된 모델을 가져오고 레이어를 시각화 하기
#모델의 제일 윗부분을 포함하고 싶지 않을 때 include_top = False
basemodel = ResNet50(weights='imagenet', include_top = False, input_tensor=Input(shape=(256,256,3))) #사전 훈련된 모델 다운
```

```python
basemodel.summary()
#2천3백만개를 최적화 해야함.
```

```python
# freezing layers in the model
#이전의 레이어 혹은 최상단 레이어를 정지하고 마지막 10개의 레이어만 훈련할 수 있도록 하려면.
for layer in basemodel.layers[:-10]:
    layers.trainable = False #가중치 훈련을 중지하라는 것
```

# TASK #7: BUILD AND TRAIN DEEP LEARNING MODEL

```python
headmodel = basemodel.output
headmodel = AveragePooling2D(pool_size=(4,4))(headmodel)
headmodel = Flatten(name = 'fatten') (headmodel)
headmodel = Dense(256, activation='relu')(headmodel)
headmodel = Dropout(0.3)(headmodel)
headmodel = Dense(256, activation='relu')(headmodel) #128에서 256으로 레이어 추가
headmodel = Dropout(0.3)(headmodel) #드롭아웃 0.2에서 0.3으로 더하기
headmodel = Dense(4, activation = 'softmax')(headmodel)

#레이어 수정후 아래의 모델코드들도 재실행
```

```python
model = Model(inputs= basemodel.input, outputs = headmodel)
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer=optimizers.RMSprop(lr=1e-4), metrics=['accuracy'])
```

```python
"""
모델을 훈련할 때 훈련 데이터와 교차 검증 데이터 모두에서 오류가 발생하면 둘 다 실제로 감소하는지를 확실하게 하고,
둘 다 감소한다면 이 모델이 일반화 될 수 있음. 
그러나 훈련 오류는 감소하는데 유효성(validation)은 올라간다면 이건 모델이 훈련 데이터의 모든 세부 사항을 배우려고 하는것을 의미
따라서 훈련 데이터에 과적합 되어 일반화에 실패

이걸 방지하기 위해 earlystopping 사용
"""
# using early stopping to exit training if validation loss is not decreasing after certain number of epochs (patience)
#mode='min' 이면 유효손실 모니터
#훈련을 중단하기로 결정하기 전 특정 수의 에포크에 가중치를 줘야함,
#조기 중지전 20개의 에포크 기다림
earlystopping = EarlyStopping(monitor = 'val_loss', mode= 'min', patience=20)

# save the best model with lower validation loss
# hdf5파일은 모든 지식, 모든값, 가중치를 포함하는 하나의 파일, 이 파일을 배포하고 연습 가능
# verbose=1이면 학습이 되는 모습을 볼 수 있음
# save_best_only=True 최소한의 검증 손실이 있는 최선의 모델만 하나 저장 
checkpointer = ModelCheckpoint(filepath='weights.hdf5', verbose=1, save_best_only=True)
```

```python
training_generator = image_generator.flow_from_directory(batch_size=4, directory=X_Ray_Directory, shuffle=True, target_size =(256,256), class_mode = 'categorical', subset='training')
validation_generator = image_generator.flow_from_directory(batch_size=4, directory=X_Ray_Directory, shuffle=True, target_size =(256,256), class_mode = 'categorical', subset='validation')

#네가지 종류에 속하는 428개의 이미지를 찾았고 104개의 이미지는 검증을 위한 다른 네 종류에 속함
```

```python
#훈련데이터에 모델을 맞춤
#model.fit_generator: 모델을 훈련 제너레이터로 훈련
#steps_per_epoch : 에포크는 많을수록 좋음
#callbacks: 언제 멈춰야하는지 입력
history = model.fit_generator(train_generator, epochs=5, validation_data=validation_generator, callbacks=[checkpointer, earlystopping])
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

prediction = []
original = []
image = []

for i in range(len(os.listdir(test_directory))):
    for item in os.listdir(os.path.join(test_directory, str(i))): #루프
        img = cv2.imread(os.path.join(test_directory, str(i), item)) #아이템을 읽을 때마다 이미지 리사이즈
        img = cv2.resize(img,(256,256)) #리사이즈
        image.append(img) #이미지 추가
        img = img /255 #이미지 정규화
        img = img.reshape(-1, 256, 256, 3) #이미지 변형
        predict = model.predict(img) #이미지를 훈련모델에 넣고 
        predict = np.argmax(predict) #argmax로 실제 예측값 산출
        prediction.append(predict) 
        original.append(i)
```

```python
score = accuracy_score(original, prediction)
print('test accuracy: {}'.format(score))
```

```python
L = 5
W = 5
fig, axes = plt.subplots(L, W, figsize=(12,12))
axes = axes.ravel()

for i in np.arange(0, L*W):
    axes[i].imshow(image[i])
    axes[i].set_title('Guess={}\nTrue={}'.format(str(label_names[prediction[i]]), str(label_names[original[i]])))
    axes[i].axis('off')
                      
plt.subplots_adjust(wspace = 1.2)
```

```python
print(classification_report(np.asarray(original), np.asarray(prediction)))
```

```python
cm = confusion_matrix(np.asarray(original), np.asarray(prediction))
ax = plt.subplot()
sns.heatmap(cm, annot=True, ax=ax)

ax.set_xlabel('Predicted')
ax.set_ylabel('Original')
ax.set_title('confusion_matrix')
```
