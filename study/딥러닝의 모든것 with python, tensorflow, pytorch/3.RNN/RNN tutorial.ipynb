### 기울기 소실
- 네트워크를 통해 오류를 전파할 때 풀어진 시간적루프를 거쳐야함. 그렇게 하며 자신들과 연결된 뉴런 레이어층들을 통과하게 됨. 이 숨겨진 층들은 그것들 자신과 연결되어 있고그듯들은 w rec 는 w 순환 가중치라고 불리는 가충치를 통해 연결되어있음. 이 가중치가 수업이 많은 시간에 거쳐 여러번 적용되므로 기울기를 빠르게 감소시키는 원인이 됨. 즉 맨 왼쪽에 있는 층의 가중치는 맨 오른쪽에 있는 가중치 보다 훨씬 더 느리게 업데이트 되며 도미노 효과를 생성. 왜냐하면 맨 왼쪽 층의 가중치가 매우 중요하기 때문이며 그것들은 맨 오른쪽 층에 대한 입력값인 해당 계층들의 출력을 지시하기 때문에 네트워크 전체 훈련에 애로사항을 겪는것임. 이것을 기울기 소실 문제라고 부름. W rec이 작으면(1보다 작으면) 기울기 소실이 일어나고 크면(1보다 크면) 기울기 폭주가 일어나게 됨.

### 기울기 폭주 해결방법
- 역전파 잘라내기 :특정 시점 이후로 다시 전파하는 것을 멈추는 것. 하지만 모든 가중치를 업데이트 하지 않기 때문에 최적이 아닐 수 있음. 그래도 전파하는 것을 멈추지 않으면 완전히 관련없는 네트워크를 갖게 됨.
- 패널티 : 기울기가 패널티를 받고 인위적으로 줄이는것. 
- gradient clipping : 기울기의 최대 제한값을 가질 수 있으며, 절대 기울기가 해당 값을 넘지 않도록 할 수 있음. 만약 해당 값을 초과하면 네트워크를 통해 더 아래로 전파될 때 해당 값을 그대로 유지

### 기울기 소실 해결방법
- 가중치 초기화 : 기울기가 소실될 가능성을 최소화 하기위해 가중치를 초기화
- echo state 네트워크 
- Long Short Term Memory Networks (LSTMs) : 순환 신경망을 구현하기 위한 네트워크
